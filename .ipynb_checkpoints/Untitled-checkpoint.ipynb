```python
from etherscan import Etherscan
from web3.auto.infura import w3
from ens.auto import ns
import os
import time
from dotenv import load_dotenv
load_dotenv()
```

```python
eth = Etherscan(os.environ['ETHERSCAN_API_KEY'])
```

```python
watch_list = ['0x2d407ddb06311396fe14d4b49da5f0471447d45c']
```

```python
last_block = eth.get_block_number_by_timestamp(int(time.time()), "before")
```

```python
new_updates = []
for addr in watch_list:
    try:
        new_updates.append({addr: eth.get_normal_txs_by_address(addr, -10, 'latest', 'desc')})
    except AssertionError as error:
        print(error)
```

```python
import json
for key in json.loads(json.dumps(new_updates[0])):
    print(ns.name(key))
```
